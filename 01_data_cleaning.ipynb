```python
import pandas as pd

df = pd.read_csv("amazon.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['discounted_price'] = (
    df['discounted_price']
    .replace('[₹,]', '', regex=True)
    .astype(float)
)

df['actual_price'] = (
    df['actual_price']
    .replace('[₹,]', '', regex=True)
    .astype(float)
)

df['discount_percentage'] = (
    df['discount_percentage']
    .str.replace('%', '')
    .astype(float)
)
```

```python
df['rating_count'] = (
    df['rating_count']
    .replace(',', '', regex=True)
    .fillna(0)
    .astype(int)
)
```

```python
# Clean rating column
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Fill missing/invalid ratings with median
df['rating'] = df['rating'].fillna(df['rating'].median())
```

```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
# create a new column price_gap 
df['price_gap'] = df['actual_price'] - df['discounted_price']
```

```python
df[['product_id','price_gap']].head(10)
```

```python
#create a new column rating_segment
df['rating_segment'] = pd.cut(
    df['rating'],
    bins=[0,3,4,5],
    labels=['Poor','Average','Excellent']
)
```

```python
df[['product_id','rating_segment']].head(10)
```

```python
#create a new column discount_bucket 
df['discount_bucket'] = pd.cut(
    df['discount_percentage'],
    bins=[0,20,40,60,100],
    labels=['Low','Medium','High','Very High']
)
```

```python
df[['discount_percentage','discount_bucket']].head(10)
```

```python
#create a new column revenue_score
import numpy as np

df['revenue_score'] = (
    df['discounted_price'] *
    np.log1p(df['rating_count']) *
    df['rating']
)
```

```python
df[['product_id','revenue_score']].head(10)
```
